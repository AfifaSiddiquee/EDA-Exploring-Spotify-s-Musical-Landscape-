<h3> 1.0.5&nbsp;&nbsp;&nbsp;Cleaning the Data:</h3>

```python
df.isna()
```

```python
df.isna().sum()
```

```python
# Imputing 'genre' column with the most frequent genre
df['genre'] = df['genre'].fillna(df['genre'].mode()[0])
df
```

```python
# Imputing numerical columns with the mean
numerical_columns = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 
                     'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 
                     'duration_ms', 'time_signature']
for col in numerical_columns:
    df[col] = df[col].fillna(df[col].mean())
df
```

```python
# For 'type', 'id', 'uri', 'track_href', and 'analysis_url',dropping these columns 
df = df.drop(['type', 'id', 'uri', 'track_href', 'analysis_url'], axis=1)
df
```

```python
#cleaned data
df.isna().sum()
```

<h3> 1.0.6&nbsp;&nbsp;&nbsp;Analyzing & Visualizing the Data:</h3>

#### 1. TOP 10 POPULAR ARTISTS:
The analysis identified the top 10 popular artists in the Spotify Song Attributes Dataset based on the number of tracks. The list includes artists such as blackbear, Lauv, and Linkin Park and others indicating their significant presence in the dataset.

```python
top_ten_artist=df.groupby("artistName").count().sort_values(by ="genre", ascending=False)["genre"][:10]
top_ten_artist
```

```python
top_ten_artist.plot.barh()
plt.title("TOP 10 POPULAR ARTISTS")
plt.show()
```

#### 2. TOP 10 LOUDEST TRACKS:
The analysis identified the top 10 loudest tracks based on their loudness levels.These tracks exhibit significantly high loudness levels compared to others in the dataset, providing insights into the dynamic range of music in the dataset.

```python
top_ten_loudest_track=df[["loudness","trackName"]].sort_values(by="loudness",ascending=True)[:10]
top_ten_loudest_track
```

```python
plt.figure(figsize=(20,4))
sns.barplot(x="loudness",y="trackName",data=top_ten_loudest_track)
plt.title("TOP 10 LOUDEST TRACKS")
plt.show()
```

#### 3. ARTIST WITH MOST DANCEABILITY SONGS:
The analysis revealed the artists with the most danceable songs in the Spotify Song Attributes Dataset.The findings provide insights into the danceability trends among different artists in the dataset.

```python
top_ten_artists_danceable_songs=df[["danceability","trackName","artistName"]].sort_values(by="danceability",ascending=False)[:10]
top_ten_artists_danceable_songs
```

```python
plt.figure(figsize=(20,4))
sns.barplot(x="danceability",y="artistName",data=top_ten_artists_danceable_songs)
plt.title("ARTIST WITH MOST DANCEABILITY SONGS")
plt.show()
```

#### 4. TOP 5 INSTRUMENTALNESS TRACKS:
The analysis  identified the top 5 tracks with the highest instrumentalness. These tracks are characterized by their lack of vocal content and focus on instrumental elements. These findings highlight the presence of instrumental music in the dataset and provide insights into the diversity of musical styles represented.

```python
top_five_instrumental_tracks=df[["instrumentalness","trackName","artistName"]].sort_values(by="instrumentalness",ascending=False)[:5]
top_five_instrumental_tracks
```

```python
plt.figure(figsize=(12,7))
plt.pie(x="instrumentalness",data=top_five_instrumental_tracks,autopct='%1.2f%%',labels=top_five_instrumental_tracks.trackName)
plt.title("TOP 5 INSTRUMENTALNESS TRACKS:")
plt.show()
```

#### 5. MULTIPLE FEATURE PLOTS
The analysis included multiple feature plots to visualize the distribution of various music attributes (tempo, loudness, acousticness, danceability, duration_ms, energy, instrumentalness, liveness, speechiness, valence) based on the mode of the tracks. The mode indicates whether a track is in a major (mode=1) or minor (mode=0) key. The plots provide insights into how these attributes differ between tracks in major and minor keys.



```python
interest_feature_cols=["tempo","loudness","acousticness","danceability","duration_ms",
                       "energy","instrumentalness","liveness","speechiness","valence"]
```

```python
for feature_col in interest_feature_cols:
    pos_data = df[df["mode"] == 1][feature_col]
    neg_data = df[df["mode"] == 0][feature_col]

    plt.figure(figsize=(9, 3.3))  # Create a new figure for each plot
    sns.histplot(pos_data, bins=30, label="Positive", color="green", alpha=0.6, kde=True)
    sns.histplot(neg_data, bins=30, label="Negative", color="red", alpha=0.6, kde=True)

    plt.legend(loc="upper right")
    plt.title(f"Positive And Negative Histogram Plot For {feature_col}")
    plt.show()
```
